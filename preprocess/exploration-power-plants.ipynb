```python
def plot_pie_chart(df, column):
    counts = df[column].value_counts()

    threshold = 0.03  # 3%
    total = counts.sum()
    counts_clean = counts[counts / total >= threshold]
    counts_clean["Other"] = counts[counts / total < threshold].sum()

    plt.figure(figsize=(8, 8))
    plt.pie(
        counts_clean,
        labels=counts_clean.index,
        autopct="%1.1f%%",
        colors=plt.cm.tab20.colors,
    )
    #plt.title("Power Plants by " + column.replace('_', ' ').title())
    # increase font size for percentage and name
    #plt.tight_layout()
plt.show()
```

```python
plot_pie_chart(df, "country")
```

```python
plot_pie_chart(df, "primary_fuel")
```
